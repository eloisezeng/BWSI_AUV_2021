# Detect Buoys using HSV

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
frame_number = 14
img = cv2.imread(f'buoy_simulation/frame_{frame_number:02d}.jpg')

imhsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
imhsv = cv2.boxFilter(imhsv, -1, (5, 5))
plt.imshow(imhsv)
plt.show()
```

```python
filter_size = (10, 10)

hfilt = cv2.boxFilter(imhsv[:, :, 0], cv2.CV_32F, filter_size)
sfilt = cv2.boxFilter(imhsv[:, :, 1], cv2.CV_32F, filter_size)
vfilt = cv2.boxFilter(imhsv[:, :, 2], cv2.CV_32F, filter_size)
fig, ax = plt.subplots(1, 3, figsize=(15, 6))
ax[0].imshow(hfilt, cmap='hsv')
ax[1].imshow(sfilt, cmap='Blues')
ax[2].imshow(vfilt, cmap='Greys')
plt.show()
```

# RGB appears more effective for detecting red and green buoys than HSV
